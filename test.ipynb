##<font size="5" color="black" face="Times"><b>&emsp;3.1. Разбор теоретических вопросов</b></font><a name = 'part31'>

<font size="3" color="black" face="Times"><div align="justify">&emsp;
* ```Git``` - это консольная утилита, для отслеживания и ведения
истории изменения файлов, в вашем проекте. Чаще всего его используют для кода, но можно и для других файлов. Например, для картинок - полезно для дизайнеров.
С помощью Git-a вы можете откатить свой проект до более старой версии, сравнивать, анализировать или сливать свои изменения в репозиторий.
Репозиторием называют хранилище вашего кода и историю его изменений. Git работает локально и все ваши репозитории хранятся в определенных папках на жестком диске.
<font size="3" color="black" face="Times"><div align="justify">&emsp;
* ```GitHub``` можно считать соцсетью для хранения кода. Это настоящая Мекка для технарей. Здесь вы можете попрактиковаться в разработке и придумать что-то свое, найти множество open-source проектов, передовых технологий, различных функций и дизайнов.
На GitHub вы учитесь и участвуете в других проектах, храните код для работы или учебы, берете код других проектов и вникаете во все детали.
<font size="3" color="black" face="Times"><div align="justify">&emsp;
* ```Software versioning``` — это процесс создания уникальных имен или номеров для различных версий продуктов программного обеспечения.

##<font size="5" color="black" face="Times"><b>&emsp;4.1. Задание</b></font><a name = 'part41'>

<font size="3" color="black" face="Times"><div align="justify">
<b>Необходимые применяемые элементы</b>
* Система контроля версий: Программа для управления версиями кода.
* Локальный репозиторий: Инициализация, отслеживание изменений.
<li>Отчет: Файл, добавление в репозиторий, фиксация изменений.
<li>Ветвление: Создание новой ветки, переключение между ветками, фиксация изменений в ветке.
<li>Слияние: Слияние изменений из ветки в основную ветку.
<li>Удаленный репозиторий: Аккаунт на платформе для хостинга кода, создание удаленного репозитория.
<li>Отправка: Связывание локального и удаленного репозитория, отправка изменений на удаленный репозиторий.
<li>Размещение статики: Публикация статического сайта с помощью сервиса хостинга.
<li>Завершение отчета: Обновление файла отчета, фиксация и отправка изменений.
<font size="3" color="black" face="Times"><div align="justify">

<b>Ожидаемые результаты выполнения работы</b>

* Установить и настроить Git.
<li>Будет создан локальный репозиторий, отчет будет добавлен и зафиксирован.
<li>Будет создана ветка, изменения в отчете будут зафиксированы в ветке.
<li>Изменения из ветки будут слиты в основную ветку.
<li>Репозиторий будет создан, локальный код будет загружен на GitHub.
<li>Статический сайт будет опубликован на GitHub
<li>Завершённый отчет со всеми описаниями будет зафиксирован в Git.
</ol>
